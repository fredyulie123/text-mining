### Example: Sentiment using NLTK Vader

 Vader is lexicon and rule-based sentiment analysis package that is specifically attuned to sentiments expressed in social media
 
 (Advanced: If you are interested to read about how Vader was developed, read this research paper: http://comp.social.gatech.edu/papers/icwsm14.vader.hutto.pdf)

```python
import pandas as pd
```

```python
# read file
df = pd.read_table("short_input.txt",header = None, names = ['Message'])
df.head()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer #Vader 

sid = SentimentIntensityAnalyzer()
for sentence in df['Message']:
    ss = sid.polarity_scores(sentence)
    for k in ss:
         print('{0}: {1},  '.format(k, ss[k]), end='')
    print()
```
