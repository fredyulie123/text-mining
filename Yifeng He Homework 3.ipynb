## sentiment Analysis

```python
import pandas as pd
```

```python
df = pd.read_csv("Fashion.csv")
df.head()
df = df.rename(columns = {'review text':'review_text'})
```

```python
def IsNotNull(value):
    return value is not None and len(value) > 0
```

```python
dict_p = []
f = open('positive-words.txt', 'r')   
for line in f:
    t = line.strip().lower()
    if (IsNotNull(t)):
        dict_p.append(t)
f.close()
```

```python
dict_n = []
f = open('negative-words.txt', 'r')   
for line in f:
    t = line.strip().lower()
    if (IsNotNull(t)):
        dict_n.append(t)
f.close()
```

```python
poscnt = []
negcnt = []
netcnt = []

for nrow in range(0,len(df)):
    
    text = df.review_text[nrow].lower()
    qa = 0
    qb = 0

    for word in dict_p :
        if (word in text) :
            qa = qa + 1

    for word in dict_n :
        if (word in text) :
            qb = qb + 1

    qc = qa - qb
    
    poscnt.append(qa)
    negcnt.append(qb)
    netcnt.append(qc)
```

```python
df['poscnt'] = poscnt
df['negcnt'] = negcnt
df['netcnt'] = netcnt
df
```

```python
df.to_csv('fashion_result.csv', index_label = 'id')
```

## Generate polarity scores

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer #Vader 
m = []
sid = SentimentIntensityAnalyzer()
for sentence in df.review_text:
    ss = sid.polarity_scores(sentence)
    for k in ss:
         print('{0}: {1},  '.format(k, ss[k]), end='')
    m.append(ss)
    print()
```

```python
m[4]
```

```python
h = pd.DataFrame(m)
```

```python
h.drop(['compound'], axis=1)
columns_titles = ['pos', 'neg', 'neu']
h = h.reindex(columns = columns_titles)
```

```python
h.to_csv('sentiment.csv', index_label='id')
```
